```python
df = pd.read_excel('AirQualityUCI.xlsx', sheet_name="AirQualityUCI_updated")
df.shape
```

```python
df[df['T'] != -200].shape
```

```python
df[df['T'] != -200].head()
```

```python
df_new = df[df['T'] != -200]
```

```python
df_new.head()
df_new.shape
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
#le = LabelEncoder()
#df_new['Time'] = le.fit_transform(df_new['Time'])
```

```python
df_new.head()
```

```python
df_new.drop(['PT08_S1_CO', "PT08_S2_NMHC", "PT08_S3_NOx", "PT08_S4_NO2", "PT08_S5_O3"], axis = 1, inplace = True)
df_new.head()
```

```python
scaler = StandardScaler()
```

```python
#df[["CO(GT)","PT08.S1(CO)","NMHC(GT)","C6H6(GT)","PT08.S2(NMHC)","NOx(GT)",'PT08.S3(NOx)','NO2(GT)','PT08.S4(NO2)','PT08.S5(O3)','T','RH']] = scaler.fit_transform(df[["CO(GT)","PT08.S1(CO)","NMHC(GT)","C6H6(GT)","PT08.S2(NMHC)","NOx(GT)",'PT08.S3(NOx)','NO2(GT)','PT08.S4(NO2)','PT08.S5(O3)','T','RH']])
```

```python
df_new.drop(["Date", "Time"], axis=1, inplace=True)
```

```python
df_new.head()
```

```python
df_new = scaler.fit_transform(df_new)
```

```python
df_new_frame = pd.DataFrame(df_new)
```

```python
df_new_frame.columns = ["CO_GT", "NMHC_GT", "C6H6_GT", "NOx_GT", "NO2_GT", "T", "RH", "AH"]
```

```python
df_new_frame.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
model = LinearRegression()
```

#### Taking "T" as target column


```python
x = df_new_frame.drop(["T", "CO_GT", "NMHC_GT"], axis = 1)
y = df_new_frame["T"]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 111)
```

```python
model.fit(xtrain, ytrain)
```

```python
predict = model.predict(xtest)
```

```python
plt.scatter(ytest, predict)
plt.show()
```

```python
print("metrics.mean_absolute_error : ", metrics.mean_absolute_error(ytest, predict))
print("metrics.mean_squared_error : ", metrics.mean_squared_error(ytest, predict))
print("np.sqrt(metrics.mean_squared_error : ", np.sqrt(metrics.mean_squared_error(ytest, predict)))
print("metrics.r2_score : ", metrics.r2_score(ytest, predict))
```

```python
import statsmodels.formula.api as smf
```

```python
est = smf.ols("T ~ C6H6_GT + NOx_GT + NO2_GT + + RH + AH", df_new_frame).fit()
est.summary()
```

#### as you can see, that CO_GT and NMHC_GT has p-value > 0.05, so we will omit it

#### Taking "AH" as target column


```python
x = df_new_frame.drop(["AH", "NO2_GT", "CO_GT"], axis = 1)
y = df_new_frame["AH"]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 111)
```

```python
model.fit(xtrain, ytrain)
```

```python
predict = model.predict(xtest)
```

```python
plt.scatter(ytest, predict)
plt.show()
```

```python
print("metrics.mean_absolute_error : ", metrics.mean_absolute_error(ytest, predict))
print("metrics.mean_squared_error : ", metrics.mean_squared_error(ytest, predict))
print("np.sqrt(metrics.mean_squared_error : ", np.sqrt(metrics.mean_squared_error(ytest, predict)))
print("metrics.r2_score : ", metrics.r2_score(ytest, predict))
```
